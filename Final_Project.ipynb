```python
"""
Data Collection for the Final Report

Dawson Dolansky
JupyterLab
4/25/2021
"""


import numpy as np
import matplotlib.pyplot as plt

vRed22, fRed22, vRedBar22, fRedBar22, vPink22, fPink22, vRedPink22, fRedPink22, vRed15, fRed15, vRedBar15, fRedBar15, vPink15, fPink15, vRedPink15, fRedPink15 = np.loadtxt("C:\\Users\Dawson Dolansky\Daily Life\James Madison University\\2nd_Year\Spring Semester\PHYS 247\Final Project\FinalProject_Sheet5.csv", skiprows = 2, unpack = True, delimiter=',')

# Fixing the issues of the data with trialing zeros
list1 = [vRed22, fRed22, vRedBar22, fRedBar22, vPink22, fPink22, vRedPink22, fRedPink22, vRed15, fRed15, vRedBar15, fRedBar15, vPink15, fPink15, vRedPink15, fRedPink15]
vR22 = []
fR22 = []
vRB22 = [] 
fRB22 = [] 
vP22 = [] 
fP22 = [] 
vRP22 = [] 
fRP22 = [] 
vR15 = [] 
fR15 = [] 
vRB15 = [] 
fRB15 = [] 
vP15 = [] 
fP15 = [] 
vRP15 = [] 
fRP15 = []
list2 = [vR22, fR22, vRB22, fRB22, vP22, fP22, vRP22, fRP22, vR15, fR15, vRB15, fRB15, vP15, fP15, vRP15, fRP15]
for i in range(0, len(list1), 1): # Lists
    for j in range(0, len(list1[i]), 1): # Values
        if (list1[i][j] != 0):
            list2[i] = np.append(list2[i], list1[i][j])            

fResonance = []
for i in range(0, 16): # Lists
    fResonance = np.append(fResonance, np.min(list2[i]))
    
print(fResonance)
resonancePoints = [0, 1, 2, 3, 4, 5, 6, 7]
resonancePoints1 = [0, 1, 2, 3]
resonancePoints2 = [0, 1, 2, 3]

resonancePoints[0] = [499.8, fResonance[0*2]]
resonancePoints[1] = [184.8, fResonance[1*2]]
resonancePoints[2] = [70.0, fResonance[2*2]]
resonancePoints[3] = [260.0, fResonance[3*2]]
resonancePoints[4] = [663.0, fResonance[4*2]]
resonancePoints[5] = [256.0, fResonance[5*2]]
resonancePoints[6] = [87.0, fResonance[6*2]]
resonancePoints[7] = [374.4, fResonance[7*2]]

resonancePoints1[0] = 499.8
resonancePoints1[1] = 260.0
resonancePoints1[2] = 184.8
resonancePoints1[3] = 70
resonancePoints2[0] = 663.0
resonancePoints2[1] = 374.0
resonancePoints2[2] = 256.0
resonancePoints2[3] = 87.0

for i in range(0, 4, 1):
    resonancePoints1[i] = 1 / ((resonancePoints1[i] * 2 * np.pi)**2)
    resonancePoints2[i] = 1 / ((resonancePoints2[i] * 2 * np.pi)**2)
    
# Our minimum points
print("Our minimum points: [Frequency, Voltage Drop]\n", resonancePoints)
    
# Interpolation
L = [0.00781, 0.02677, 0.21006, 0.01225, 0.00781, 0.02677, 0.21006, 0.01225]
L1 = [0.00781, 0.01225, 0.02677, 0.21006]
resonance = [393.69, 212.70, 75.94, 314.41, 532.76, 287.83, 102.76, 425.47]
fig1, subs = plt.subplots(nrows = 4, ncols = 1, figsize = (25, 50))
for i in range(0, 4, 1):
    subs[i].errorbar(x = list2[i*2+1], y = list2[i*2+0], fmt = 'b->', label = "22uF Circuit", xerr = .1*2*np.pi, yerr = 20)
    subs[i].errorbar(x = list2[(i+4)*2+1], y = list2[(i+4)*2+0], fmt='r->', label = "15uF Circuit", xerr = .1*2*np.pi, yerr = 20)
    subs[i].axvline(resonance[i]).set_color("b")
    subs[i].axvline(resonance[i+4]).set_color("r")
    subs[i].grid()
    subs[i].axhline()
    subs[i].axvline()
    subs[i].set_title(f"Driving Voltage Frequency vs Current Magnitude at {L[i]} H", fontsize = 20.0)
    subs[i].set_xlabel("Driving Voltage Frequency (Hz)", fontsize = 20.0)
    subs[i].set_ylabel("Voltage Drop across the Inductor-Capacitor (mV)", fontsize = 20.0)
    subs[i].legend(fontsize = 50, loc = "lower right")

def Z(L, f, fo):
    return L*np.abs(((f*np.pi*2)**2-(fo*np.pi*2)**2)/(f*np.pi*2))

def zUnc(L, w, wo):
    return np.sqrt((.00001/L)**2 + ((.1*2*np.pi)/w)**2 + ((.1*2*np.pi)/wo)**2)

def iUnc(v, zUnc):
    return np.sqrt((.02/v)**2 + (zUnc)**2)
    
impedanceList = [[], [], [], [], [], [], [], [], []]
currentList = [[], [], [], [], [], [], [], [], []]
currentUncertainty = [[], [], [], [], [], [], [], [], []]
zUncert = 0
# Calculating the impedance for every level
for i in range(0, 4, 1): # Lists
    # Calculating the Impedance
    for j in range(0, len(list2[i*2+1]), 1):
        # If we are at the resonance, we can say that the resonance frequency we calculated was within half the intervals between frequencies
        if (list2[(i)*2+1][j] == resonancePoints[i][0]): # Z(w) = l(w^2-wo^2)/w
            impedanceList[i] = np.append(impedanceList[i], Z(L[i], list2[(i)*2+1][j]-(list2[(i)*2+1][1] - list2[(i)*2+1][0])/2, resonancePoints[i][0]))
            impedanceList[i] = np.append(impedanceList[i], Z(L[i], list2[(i)*2+1][j]+(list2[(i)*2+1][1] - list2[(i)*2+1][0])/2, resonancePoints[i][0]))
        else:
            impedanceList[i] = np.append(impedanceList[i], Z(L[i], list2[(i)*2+1][j], resonancePoints[i][0]))
        zUncert = zUnc(L[i], list2[(i)*2+1][j], resonancePoints[i][0])
        currentList[i] = np.append(currentList[i], list2[(i)*2][j] / impedanceList[i][j])
        currentUncertainty[i] = np.append(currentUncertainty[i], iUnc(list2[(i)*2][j], zUncert))
    for j in range(0, len(list2[(i+4)*2+1]), 1):
        if (list2[(i+4)*2+1][j] == resonancePoints[i+4][0]):
            impedanceList[i+4] = np.append(impedanceList[i+4], Z(L[i+4], list2[(i+4)*2+1][j]-(list2[(i+4)*2+1][1] - list2[(i+4)*2+1][0])/2, resonancePoints[i+4][0]))
            impedanceList[i+4] = np.append(impedanceList[i+4], Z(L[i+4], list2[(i+4)*2+1][j]+(list2[(i+4)*2+1][1] - list2[(i+4)*2+1][0])/2, resonancePoints[i+4][0]))
        else:
            impedanceList[i+4] = np.append(impedanceList[i+4], Z(L[i+4], list2[(i+4)*2+1][j], resonancePoints[i+4][0]))
        zUncert = zUnc(L[i+4], list2[(i+4)*2+1][j], resonancePoints[i+4][0])
        currentList[i+4] = np.append(currentList[i+4], list2[(i+4)*2][j] / impedanceList[i+4][j]) # V / Z
        currentUncertainty[i+4] = np.append(currentUncertainty[i+4], iUnc(list2[(i+4)*2][j], zUncert))

# Plotting current vs frequency
fig2, sub = plt.subplots(nrows = 4, ncols = 1, figsize = (25, 50))
for i in range(0, 4, 1):
    sub[i].errorbar(x = list2[i*2+1], y = currentList[i], fmt = 'b->', label = "22uF Circuit", xerr = (.1*2*np.pi), yerr = currentUncertainty[i])
    sub[i].errorbar(x = list2[(i+4)*2+1], y = currentList[i+4], fmt='r->', label = "15uF Circuit",  xerr = (.1*2*np.pi), yerr = currentUncertainty[i+4])
    sub[i].axvline(resonance[i]).set_color("b")
    sub[i].axvline(resonance[i+4]).set_color("r")
    sub[i].grid()
    sub[i].axhline()
    sub[i].axvline()
    sub[i].set_title(f"Driving Voltage Frequency vs Current at {L[i]} H", fontsize = 20.0)
    sub[i].set_xlabel("Driving Voltage Frequency (Hz)", fontsize = 20.0)
    sub[i].set_ylabel("Current through the Circuit (mV)", fontsize = 20.0)
    sub[i].legend(fontsize = 50, loc = "lower right")
    
# Calculating the average uncertainty of the current
averageUncertainty = 0
length = 0
for i in range(0, 4):
    for j in range(0, len(currentUncertainty[i])):
        averageUncertainty += currentUncertainty[i][j]
        length += 1
    for k in range(0, len(currentUncertainty[i+4])):
        averageUncertainty += currentUncertainty[i+4][k]
        length += 1
averageUncertainty /= length

plt.show()
fig2, sub1 = plt.subplots(nrows = 1, ncols = 1, figsize = (15, 10))


plt.plot(L1, resonancePoints1, 'b->', label = "22uF Circuit")
plt.plot(L1, resonancePoints2, 'r->', label = "15uF Circuit")
plt.grid()
plt.axhline()
plt.axvline()
plt.title(f"Inductance vs inverse square of angualar frequency", fontsize = 20.0)
plt.xlabel("Inductance (1 +/- .00001H)", fontsize = 20.0)
plt.ylabel("1 / w^2 (s^2/rad^2)", fontsize = 20.0)
plt.legend(fontsize = 50, loc = "lower right")

plt.show()
print(averageUncertainty)
```
